```python
import pandas as pd
from alpha_research.SingleAssetResearch import SingleAssetResearch
import yfinance as yf
```

```python
hsi = yf.Ticker("^HSI")
hist = hsi.history(period="max")
hist = hist[-3000:]
hist.columns = [c.lower() for c in hist.columns]
hist
```

```python
from alpha_research.factor_transformation import normalize_factor
def alpha(df, lag_1, lag_2):   
    factor = df['close'].rolling(lag_2).mean() - df['close'].rolling(lag_1).mean()
    factor.fillna(0, inplace=True)
#     print(factor.mean())
#     print(factor.std())
#     print((factor - factor.mean()) /factor.std()) 
    return normalize_factor(factor)
```

```python
factor_study = SingleAssetResearch(hist)
```

```python
factor_study.calculate_factor(alpha, lag_1=20, lag_2=30)
```
