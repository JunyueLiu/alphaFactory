```python
import pandas as pd
from alpha_research.SingleAssetResearch import SingleAssetResearch
from alpha_research.MultiAssetResearch import MultiAssetResearch
import yfinance as yf
```

```python
hsi = yf.Ticker("^HSI")
hist = hsi.history(period="max")
hist = hist[-3000:]
hist.columns = [c.lower() for c in hist.columns]
hist
```

```python
from alpha_research.factor_transformation import normalize_factor
def alpha(df, lag_1, lag_2):   
    factor = df['close'].rolling(lag_2).mean() - df['close'].rolling(lag_1).mean()
    factor.fillna(0, inplace=True)
    return normalize_factor(factor)
```

```python
#  inital SingleAssetResearch object
# if you only initial with a single dataframe, witth automatically split it into in_sample and 
factor_study = SingleAssetResearch(hist) # hist is history data of HSI index
```

```python
# send your alpha function and your alpha parameter into the ".calculate_factor" function
factor_study.calculate_factor(alpha, lag_1=10, lag_2=5)
```

```python
# now you can evaluate your alpha in sample
factor_study.set_factor_name('ma5-ma10 factor')
factor_study.evaluate_alpha()
```
